### Task:
Given a review text, count how many times each word appears (ignore case).

---

### Example:
**Input:**  
"This product is good and this service is excellent"  

**Output:**  
{'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

```python
# Word frequency counter that prints dictionary-like output, WITHOUT using dicts
text = input("Enter your review text: ").strip().lower()

# simple tokenization (split on whitespace)
tokens = text.split()

words = []   # unique words encountered
counts = []  # parallel counts

for t in tokens:
    if t in words:
        idx = words.index(t)
        counts[idx] += 1
    else:
        words.append(t)
        counts.append(1)

# build a string that looks like a dict: {'word': count, ...}
pairs = []
for i in range(len(words)):
    # ensure words are quoted with single quotes like a real dict repr
    pairs.append(f"'{words[i]}': {counts[i]}")

dict_like = "{" + ", ".join(pairs) + "}"
print(dict_like)
```
